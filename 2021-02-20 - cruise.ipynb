## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_location = "/FileStore/tables/cruise_ship_info-4.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# Transform data structure
from pyspark.ml.linalg import Vectors
df_new = df.rdd.map(lambda x: [Vectors.dense(x[2:8]), x[-1]]).toDF(['features', 'Label'])
df_new.show(5)
```

```
# Build linear regression model 1
 
from pyspark.ml.regression import LinearRegression
lr1 = LinearRegression(featuresCol = 'features', labelCol = 'Label')
# Fit the model
lrmodel = lr1.fit(df_new)
# Predictions 1
pred1 = lrmodel.transform(df_new)
pred1.show(5)
```

```
display(lrmodel, df_new)
```

```
# Module evaluation 1
# Root Mean Square Error
from pyspark.ml.evaluation import RegressionEvaluator 
evaluator1 = RegressionEvaluator(predictionCol='prediction', labelCol='Label')
evaluator1.setMetricName('rmse').evaluate(pred1)
evaluator1.setMetricName('r2').evaluate(pred1)
evaluator1.setMetricName('mse').evaluate(pred1)
print("R2: {}".format(evaluator1.setMetricName('r2').evaluate(pred1)))
print("MSE: {}".format(evaluator1.setMetricName('mse').evaluate(pred1)))
print("RMSE: {}".format(evaluator1.setMetricName('rmse').evaluate(pred1)))
```

```
print("Coefficients: {} Intercept: {}".format(lrmodel.coefficients,lrmodel.intercept))
```

```
# Another way to transfer data structure
```

```
# Category indexing with StringIndexer
# Change the Column "Cruise_Line" to Categorical Variable
from pyspark.ml.feature import StringIndexer 
indexer = StringIndexer(inputCol='Cruise_line',outputCol='cruiseline_category')
indexed = indexer.fit(df).transform(df)
indexed.printSchema()
```

```
indexed.head(3)
```

```
#combine all the features in one single feature vector.
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=['Age',
 'Tonnage',
 'passengers',
 'length',
 'cabins',
 'passenger_density',
 'cruiseline_category'],outputCol = 'features')
output = assembler.transform(indexed)
```

```
ad_df= output.select('features','crew')
ad_df.show()
```

```
# Build linear regression model 2
 
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(featuresCol = 'features', labelCol = 'crew')
# Fit the model 2
lr_model = lr.fit(ad_df)
# Make Prediction 2
pred = lr_model.transform(ad_df)
pred.show(5)
```

```
# Module evaluation 2
# R Squared 2
from pyspark.ml.evaluation import RegressionEvaluator 
evaluator = RegressionEvaluator(predictionCol='prediction', labelCol='crew')
evaluator.setMetricName('r2').evaluate(pred)
print("R2: {}".format(evaluator.setMetricName('r2').evaluate(pred)))
```

```
# Module evaluation 2
# Root Mean Square Error 2
from pyspark.ml.evaluation import RegressionEvaluator 

evaluator = RegressionEvaluator(predictionCol='prediction', labelCol='crew')
evaluator.setMetricName('rmse').evaluate(pred)
print("RMSE: {}".format(evaluator.setMetricName('rmse').evaluate(pred)))
```

```
# Module evaluation 2
# Mean Square Error 2
evaluator = RegressionEvaluator(predictionCol='prediction', labelCol='crew')
evaluator.setMetricName('mse').evaluate(pred)
print("MSE: {}".format(evaluator.setMetricName('mse').evaluate(pred)))
```

```
print("Coefficients: {} Intercept: {}".format(lr_model.coefficients,lr_model.intercept))
```
